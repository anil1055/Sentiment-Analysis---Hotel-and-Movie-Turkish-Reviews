# Creating Positive-Negative Word's List and Listing Most Common Words

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')

pos_countList = []
neg_countList = []
ind = 0 
#Calculate most common words according to positive-negative labels
for test in trains:
  text_tokens = word_tokenize(test)
  for token in text_tokens:
    if ind <13350: neg_countList.append(token) #2900 for hotel, 13350 for movie
    else: pos_countList.append(token)
  ind += 1

from collections import Counter
negcounts = Counter(neg_countList)
poscounts = Counter(pos_countList)
negtemp = list(negcounts.items()) 
postemp = list(poscounts.items()) 
print(negcounts.most_common(20))
print(poscounts.most_common(20))
```

**Updating Dataset according to Removing First "k" common words for opposite sentiment**

```python
negRemove = []
posRemove = []
for key,value in negtemp:
  if value>1250 : # 1240 for hotel, 1600 movie and (key != 'iyi' or key != 'güzel') for movie (k-5: 1700, k-15: 940 hotel, k-5:3000 ,k-15: 1250 movie)
    negRemove.append(key)
for key,value in postemp:
  if value>1120: # 496 for hotel, 1400 for movie (k-5: 800, k-15: 280 hotel, k-5:2538 ,k-15: 1120 movie)
    posRemove.append(key)

ind = 0
testV2 = []
trainV2 = []
total_word = []
for test in tests:
  text_tokens = word_tokenize(test)
  without = []
  if ind<13350: without = [word for word in text_tokens if not word in posRemove] #2900 for hotel, 13350 for movie
  else: without = [word for word in text_tokens if not word in negRemove]
  for token in without:
    total_word.append(token)
  filtered_sentence = (" ").join(without)
  testV2.append(filtered_sentence)
  ind += 1
print("Test file updated.")
ind = 0 
for test in trains:
  text_tokens = word_tokenize(test)
  without = []
  if ind<13350: without = [word for word in text_tokens if not word in posRemove] #2900 for hotel, 13350 for movie
  else: without = [word for word in text_tokens if not word in negRemove]
  for token in without:
    total_word.append(token)
  filtered_sentence = (" ").join(without)
  trainV2.append(filtered_sentence)
  ind += 1
print("Train file updated.")
print("Toplam kelime sayısı (Total word count): " + str(len(total_word)))
total_words = list(set(total_word))
print("Eşsiz kelime sayısı (Unique word count): " + str(len(total_words)))
```

**Editing csv File and model parameter**

```python
import pandas as pd

df=pd.read_csv("movie15.csv") #tweets.csv
df.columns=["labels","text"]
#categorized labels
df.labels=pd.Categorical(df.labels)
df['labels'] = pd.factorize(df.labels)[0]
```

```python
# model
from simpletransformers.classification import ClassificationModel
import torch,sklearn

#train-eval splitting
train_df=df[:26700] #26700 for movie, 5800 for hotel
eval_df=df[26700:]
```

# Training Language Models

```python
model = ClassificationModel('bert', 'bert-base-multilingual-uncased', num_labels=2, 
                            args={'reprocess_input_data': True, 'overwrite_output_dir': True, 'num_train_epochs': 3})

model.train_model(train_df, acc=sklearn.metrics.accuracy_score)
```

Training all models for Turkish sentiment analysis

```python
model_args = {
    "use_early_stopping": True,
    "early_stopping_delta": 0.01,
    "early_stopping_metric": "mcc",
    "early_stopping_metric_minimize": False,
    "early_stopping_patience": 5,
    "evaluate_during_training_steps": 1000,
    "fp16": False,
    'overwrite_output_dir': True,
    "num_train_epochs":3
}

model_b = ClassificationModel("bert", "dbmdz/bert-base-turkish-cased",args=model_args, num_labels=2)
model_b.train_model(train_df, acc=sklearn.metrics.accuracy_score)

model_bm = ClassificationModel("bert", "bert-base-multilingual-uncased",args=model_args, num_labels=2)
model_bm.train_model(train_df, acc=sklearn.metrics.accuracy_score)

model_db = ClassificationModel("distilbert", "dbmdz/distilbert-base-turkish-cased",args=model_args, num_labels=2)
model_db.train_model(train_df, acc=sklearn.metrics.accuracy_score)

model_el = ClassificationModel("electra", "dbmdz/electra-base-turkish-cased-discriminator",args=model_args, num_labels=2)
model_el.train_model(train_df, acc=sklearn.metrics.accuracy_score)

model_alb = ClassificationModel("albert", "loodos/albert-base-turkish-uncased",args=model_args, num_labels=2)
model_alb.train_model(train_df, acc=sklearn.metrics.accuracy_score)
```

# Testing Language Models

```python
test = eval_df #df[5800:] #26700 for movie, 5800 hotel
result, model_outputs, wrong_predictions = model_b.eval_model(test)
predictions = model_outputs.argmax(axis=1)
actuals = test.labels.values
from sklearn.metrics import accuracy_score
print(accuracy_score(actuals, predictions))

test = eval_df #df[5800:] #26700 for movie, 5800 hotel
result, model_outputs, wrong_predictions = model_bm.eval_model(test)
predictions = model_outputs.argmax(axis=1)
actuals = test.labels.values
from sklearn.metrics import accuracy_score
print(accuracy_score(actuals, predictions))

test = eval_df #df[5800:] #26700 for movie, 5800 hotel
result, model_outputs, wrong_predictions = model_db.eval_model(test)
predictions = model_outputs.argmax(axis=1)
actuals = test.labels.values
from sklearn.metrics import accuracy_score
print(accuracy_score(actuals, predictions))

test = eval_df #df[5800:] #26700 for movie, 5800 hotel
result, model_outputs, wrong_predictions = model_el.eval_model(test)
predictions = model_outputs.argmax(axis=1)
actuals = test.labels.values
from sklearn.metrics import accuracy_score
print(accuracy_score(actuals, predictions))

test = eval_df #df[5800:] #26700 for movie, 5800 hotel
result, model_outputs, wrong_predictions = model_alb.eval_model(test)
predictions = model_outputs.argmax(axis=1)
actuals = test.labels.values
from sklearn.metrics import accuracy_score
print(accuracy_score(actuals, predictions))
```
